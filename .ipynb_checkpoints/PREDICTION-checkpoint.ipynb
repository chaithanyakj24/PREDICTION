```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv("dataset.csv")

# Categorize SSLC marks, PUC marks, and CGPA into bins
data['BINNED_SSLC_MARKS'] = pd.cut(data['SSLC_MARKS'], bins=[0, 60, 70, 80, 90, 100], labels=['Below 60', '60-70', '70-80', '80-90', '90-100'])
data['BINNED_PUC_MARKS'] = pd.cut(data['PUC_MARKS'], bins=[0, 60, 70, 80, 90, 100], labels=['Below 60', '60-70', '70-80', '80-90', '90-100'])
data['BINNED_CGPA'] = pd.cut(data['CGPA'], bins=[0, 6, 7, 8, 9, 10], labels=['Below 6', '6-7', '7-8', '8-9', '9-10'])

# Group by each categorical variable and calculate counts of placed individuals
categorical_variables = ['REGION', 'INTERNSHIP', 'BRANCH', 'BINNED_SSLC_MARKS', 'BINNED_PUC_MARKS', 'BINNED_CGPA']

# Set up subplots with 2 rows and 3 columns
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))

# Flatten the axes array to iterate over each subplot
axes = axes.flatten()

for i, category in enumerate(categorical_variables):
    placed_counts = data[data['STATUS'] == 'YES'].groupby(category).size()
    categories = placed_counts.index

    # Plotting pie chart with adjusted parameters
    axes[i].pie(placed_counts, labels=categories, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'Placed Candidates by {category}')
    axes[i].axis('equal')  # Ensures the pie chart is drawn as a circle
    plt.setp(axes[i].get_xticklabels(), fontsize=8)  # Adjust label font size

plt.tight_layout()
plt.show()
```
