```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv("new.csv")

# Categorize SSLC marks, PUC marks, and CGPA into bins
data['SSLC_MARKS'] = pd.cut(data['SSLC_MARKS'], bins=[0, 60, 70, 80, 90, 100], labels=['Below 60', '60-70', '70-80', '80-90', '90-100'])
data['PUC_MARKS'] = pd.cut(data['PUC_MARKS'], bins=[0, 60, 70, 80, 90, 100], labels=['Below 60', '60-70', '70-80', '80-90', '90-100'])
data['CGPA'] = pd.cut(data['CGPA'], bins=[0, 6, 7, 8, 9, 10], labels=['Below 6', '6-7', '7-8', '8-9', '9-10'])

# Group by each categorical variable and calculate counts of individuals
categorical_variables = ['REGION', 'INTERNSHIP', 'BRANCH', 'SSLC_MARKS', 'PUC_MARKS', 'CGPA']

plt.figure(figsize=(18, 12))

for i, category in enumerate(categorical_variables):
    placed_counts = data[data['STATUS'] == 'Placed'].groupby(category).size()
    total_counts = data.groupby(category).size()
    percentages = (placed_counts / total_counts) * 100
    categories = placed_counts.index

    # Plotting bar chart
    plt.subplot(2, 3, i+1)
    plt.bar(categories, percentages, color='skyblue')
    plt.ylabel('Percentage (%)')
    plt.xticks(rotation=45, ha='right')

    # Adding percentage labels above each bar
    for j, percentage in enumerate(percentages):
        plt.text(j, percentage + 1, f"{percentage:.1f}%", ha='center')

    # Adding figure tag below each subplot
    plt.text(0.5, -0.2, f"Figure 6.2.{i+1}: Placed Candidates by {category}", transform=plt.gca().transAxes, ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```
